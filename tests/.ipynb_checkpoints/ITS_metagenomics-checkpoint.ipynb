```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

## Some Exploratory Data Analysis on taxanomic Information gotten from ITS Sequencing
Sequencing was done through Laragen 16s ITS rRNA sequencing  using MiSeq, and analysis via Illumina BaseSpace App, 16s Metagenomics App

```python
taxa_df = pd.read_csv('..\\data\\ITS_sequencing\\ITS_taxanomic_summary.csv')
taxa_df.head()
```

```python
total_hits = sum(taxa_df['num_hits'])
genuses = taxa_df['Genus'].unique()
genuses = np.delete(genuses,0)
species = taxa_df['Species']

print('Identified %s unique genuses and %.2f unique species'%(len(genuses),len(species)))
```

```python
greatest_hits = taxa_df[taxa_df['%_hits'] > 2][['Genus','num_hits']]
greatest_hits.loc[len(greatest_hits)] = ['Other',total_hits - sum(greatest_hits['num_hits'])]
greatest_hits.at[4,'Genus'] = ' Ascomycota'

# Pie Chart of overall Genus Composition
colors=cm.Set3(np.arange(10))
plt.figure(figsize = (20,20))
plt.rc('font', size=40) 
plt.pie(greatest_hits['num_hits'], labels=greatest_hits['Genus'],autopct='%1.1f%%',colors=colors,pctdistance=.7)
plt.savefig('pictures\\genus_pie.png',transparent=True)
plt.show()
```

## Wood Rot Composition
Source https://www.messiah.edu/Oakes/fungi_on_wood//Taxonomic%20Classification.htm

```python
wood_rot_txt = open('..\\data\\ITS_sequencing\\wood_decay_species.txt', 'r')
wood_rot_species = wood_rot_txt.readlines()
wood_rot_txt.close()
wood_rot_species = [rot.replace('\n','') for rot in wood_rot_species]
wood_rot_species = pd.DataFrame([rot.split(' ') for rot in wood_rot_species]).drop(columns=2).rename(columns={0:'Genus',1:'Species'})
wood_rot_species.head()
```

```python
wood_rot_genuses = wood_rot_species['Genus'].to_list()
possible_rot = taxa_df[taxa_df['Genus'].isin(wood_rot_genuses)]
rot_percent = sum(possible_rot['%_hits'])
print('We have %s genuses that are possibly electrogenic' %len(possible_rot))
print('Taking % reads = % population, we have', rot_percent, '% of population as possibly electrogenic')
possible_rot
```

```python
wood_rot = possible_rot[['Genus','num_hits']].groupby('Genus').sum(axis=1).reset_index()
wood_rot_total_hits = sum(wood_rot['num_hits'])
wood_rot_greatest_hits = wood_rot[wood_rot['num_hits'] > wood_rot_total_hits * .01] 
wood_rot_greatest_hits = wood_rot_greatest_hits.append({'Genus':'Other','num_hits':wood_rot_total_hits - sum(wood_rot_greatest_hits['num_hits'])},ignore_index=True)

plt.figure(figsize = (20,20))
colors = cm.tab10([5,6,7,3,8])
plt.rc('font', size=30) 
plt.pie(wood_rot_greatest_hits['num_hits'], labels=wood_rot_greatest_hits['Genus'],colors=colors,autopct='%1.1f%%', pctdistance=.8)
plt.savefig('pictures\\wood_rot_genus_pie.png',transparent=True)
plt.show()
```
